```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
# to drop rows having value setosa
df = df[df["species"] != "setosa"]

df.head()
```

```python
df["species"] = df["species"].map({'versicolor':0 , 'virginica': 1})

#replaces values
df.head()
```

```python
# Logistic regression is used in classification problem statements

from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split

X = df.iloc[: , :-1]
y = df.loc[:, 'species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300], 'solver': ['saga'], 'l1_ratio': [0.5] }

model = GridSearchCV(estimator= LogisticRegression(), param_grid=parameter,scoring='accuracy',cv=5)

model.fit(X_train, y_train)

pred = model.predict(X_test)

```

```python
print(model.best_params_)

print(model.best_score_)
```

```python
# to find accuracy of the model
from sklearn.metrics import accuracy_score

score=accuracy_score(pred,y_test)
print(score)
```
