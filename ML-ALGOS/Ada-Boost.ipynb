```python
#steps of ada-boost are
# 1. Initialize the sample weights for all training samples equally.
# 2. Train a weak classifier on the weighted training data.
# 3. Evaluate the weak classifier's performance and calculate its error rate using formula: TE = sum(weights of misclassified samples) / sum(weights of all samples) and performance of stump =1/2 ln(1-TE/TE)
# 4. Calculate the classifier's weight based on its error rate.
# 5. Update the weights of the training samples: increase the weights of misclassified samples and decrease the weights of correctly classified samples.in this amnner when the loop runs , incorrect samples will have more chances of getting selected for training the next weak classifier.
# 6. Normalize the weights so that they sum to 1.
# 7. Run a for loop which randomly choses between 0-1 and hence choose the weak classifier for next model
# 8. Repeat steps 2-6 for a specified number of iterations or until a stopping criterion is met.
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the iris dataset
X, y = load_iris(return_X_y=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an AdaBoost classifier with decision tree stumps as weak learners
ada_boost = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=20),  # Stump (depth=1)
    n_estimators=50,
    learning_rate=1.0,
    random_state=42
)

# Train the model
ada_boost.fit(X_train, y_train)

# Make predictions
y_pred = ada_boost.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"AdaBoost Accuracy: {accuracy:.4f}")
```
