##### Bar Chart

```python
categories= np.array(["grains","fruit","veges","protien","dairy","sweets"])

values = np.array([4,3,2,5,3,1])

plt.bar(categories,values , color ="orange")
# plt.barh(categories,values , color ="orange")   ---- for horizontal bar


plt.title("Daily Consumption")

plt.show()
```

##### Pie Chart

```python
categories= np.array(["grains","fruit","veges","protien","dairy","sweets"])

values = np.array([4,3,2,5,3,1])

colors = ["red","yellow","blue", "green" , "purple", "teal"]

plt.pie( values , labels = categories,
                           autopct="%1.1f%%",
                           colors=colors,
                           explode =  [0,0,0,0,0.5,0.2],
                           shadow= True,
                           startangle= 30     )

plt.show()
```

##### Histograms

```python
scores = np.random.normal(loc=80, scale =50, size =100)

# loc descirbes median of our data
# scale tells how far data will deviate from centre, standard deviation 

scores= np.clip(scores, 0 ,100)   
#it will clip scores below 0 and above 100

plt.hist(scores,bins=40,
                color= "lightgreen",
                edgecolor= "black")

plt.show()
```
