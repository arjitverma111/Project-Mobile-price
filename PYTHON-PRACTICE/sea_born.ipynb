```python
import seaborn as sns
import pandas as pd , numpy as np , matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()

penguins = sns.load_dataset('penguins')

penguins.head(5)
```

```python
penguins["species"].value_counts()
 
penguins["island"].value_counts()
```

```python
plt.scatter(penguins["flipper_length_mm"] , penguins['body_mass_g'] , color = "orange", s=10)
```

```python
sns.scatterplot(data= penguins , x = "flipper_length_mm", y="body_mass_g", hue="island")

```

##### Styling themes


```python
sns.set_style("ticks")  #background change krta h 
#white
#dark
#whitegrid
#darkgrid
#ticks

sns.set_context("paper")    #maker size k liye



sns.scatterplot(data= penguins , x = "flipper_length_mm", y="body_mass_g", hue="island", palette = "Set2")  # palette helps in choosing colour pallete

sns.despine( left = True)
```

##### Strip plot

```python
sns.scatterplot(data= penguins , x = "species", y="body_mass_g", hue="island", style="sex", palette = "dark", alpha =0.4)

#alpha - opacity
```

```python
sns.stripplot(data= penguins , x = "species", y="body_mass_g", hue="island",  palette = "dark", dodge= True, jitter = True)

#dodge seperates out different hues, 
#jitter randomly scatters the plot for better visibility
```

##### Swarm plot

```python
sns.swarmplot(data= penguins , x = "species", y="body_mass_g", hue="island")

#this is also used to randomly scatter data but it gives data a swarm shape instead
```

##### Histogram

```python
sns.histplot(data = penguins, x = "body_mass_g", hue ="sex", multiple= "stack")

# multiple= "stack" used to make it look better
```

##### Regression plot

```python
sns.regplot(data = penguins, x = "body_mass_g", y= "flipper_length_mm")
```

##### Line plot

```python
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "body_mass_g", y= "flipper_length_mm", hue= "sex", style ="island")
```

##### Bar plot

##### Count plot

```python
sns.countplot(data = penguins, x = "species", hue= "island")
```

##### Box plot

##### Violin plot

```python
sns.violinplot(data = penguins, x = "species", y= "body_mass_g",hue = "island")
```

##### Kde plot

```python
sns.kdeplot(data=penguins, x="body_mass_g", hue="species", fill = True)
```

##### Heatmap

```python
penguins.head()
```

```python
columns= ["bill_length_mm",	"bill_depth_mm",	"flipper_length_mm",	"body_mass_g"]

penguins[columns].corr()

```

```python
sns.heatmap(data = penguins[columns].corr(), annot = True, cmap = "Blues", vmin= -0.2)

# annot marks the values
# vmin sets the min value
```

##### Rug plot

```python
sns.rugplot(data= penguins, y="body_mass_g", hue="species", palette ="Set2", )
```

##### Pair plot

```python
sns.pairplot(data= penguins)
```

```python
sns.pairplot(data=penguins,hue="sex", palette="Set2", diag_kind="kde")
```

##### Pair Grid

```python
graph = sns.PairGrid(data=penguins,hue="sex", palette = "Set2")
graph.map_upper(sns.scatterplot)
graph.map_diag(sns.histplot)
graph.map_lower(sns.kdeplot)

# it is basically a better version of pair plot, it needs to be specified that in which triangle which type of graph should be there
```
